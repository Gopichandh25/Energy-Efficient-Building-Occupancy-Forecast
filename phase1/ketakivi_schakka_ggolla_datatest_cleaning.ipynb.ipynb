```python
no_of_duplicates = df_test[df_test.duplicated()]
print(no_of_duplicates)
```

```python
df_test.drop_duplicates(inplace = True)
```

```python
#Since There are some null values, dropping the rows with null values
null_rows = df_test.isnull().sum()
print(null_rows)
df_test = df_test.dropna(axis=0)
print(df_test.isnull().sum())
```

```python
df_test.describe()
```

```python
df_test.info()
```

```python
df_test = df_test.rename(columns = {"date":"Date"})
```

```python
dates = df_test["Date"]

def change_string_into_datetimeformat(x):
  return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

converted_dates = dates.apply(change_string_into_datetimeformat)

df_test["Date"] = converted_dates
```
