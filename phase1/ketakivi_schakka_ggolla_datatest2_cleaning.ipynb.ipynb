```python
df_test2 = df_test2.drop_duplicates()
duplicate_entries = df_test2[df_test2.duplicated()]
print(duplicate_entries)
# no more duplicate entries present in the dataframe
```

```python
zero_counts = (df_test2==0).sum()
print("the number of zero values in the dataset are: ",zero_counts)
```

```python
# change all column names to upper case for consistency
df_test2.columns = df_test2.columns.str.capitalize()
print(df_test2)
```

```python
dates2 = df_test2["Date"]

def change_string_into_datetimeformat(x):
  return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

converted_dates2 = dates2.apply(change_string_into_datetimeformat)

df_test2["Date"] = converted_dates2
df_test2
df_test2.info()
```

```python
df_test2['Temperature'] = df_test2['Temperature'].round(3)
df_test2['Light'] = df_test2['Light'].round(3)
df_test2['Co2'] = df_test2['Co2'].round(3)
df_test2['Humidity'] = df_test2['Humidity'].round(3)
df_test2['Humidityratio'] = df_test2['Humidityratio'].round(3)
df_test2
```
